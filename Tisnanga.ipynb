```python
#inspecting data sets
def inspect_multiple_data(*dfs, rows=5):
    """
    Prints the first few rows of multiple DataFrames.

    Parameters:
    *dfs: List of DataFrames to inspect.
    rows: Number of rows to display from each DataFrame (default is 5).
    """
    for i, df in enumerate(dfs):
        print(f"\n=== DataFrame {i+1} - First {rows} Rows ===")
        print(df.head(rows))
```

```python
# Assuming you have these DataFrames: ratings, movies, tags, links
inspect_multiple_data(ratings, movies, tags, links)

# To show a custom number of rows (e.g., 10 rows):
inspect_multiple_data(ratings, movies, tags, links, rows=10)
```
